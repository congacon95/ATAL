```python
def get_soup(rows, path, _id):
    try:
        dr = get.driver()
        for row in rows:
            try:
                row['SOUP'] =get.soup(row['URL'], dr)
            except:
                get.pe()
    except:
        get.pe()
    dr.quit()
```

```python
get.run_threads(get.split(rows, 15), get_soup)
```

```python
#get.save(rows, 'raw/0001P')
def ftext(e):
    return e.text.strip()
get.ft = ftext
```

```python
def find_element(val, tag='div', selector='class', parent=None):
    if parent:
        print(parent)
        return parent.find(tag,{selector:val})
    return get._SOUP.find(tag,{selector:val})
get.fe = find_element
def find_element_text(val, tag='div', selector='class', parent=None):
    if parent:
        return get.ft(parent.find(tag,{selector:val}))
    return get.ft(get._SOUP.find(tag,{selector:val}))
get.fet = find_element_text
```

```python
def fpf(text):
    return text.replace('\t', '').replace('\n', '').replace(' ','').replace('.', ' ').replace('P', '').replace('F', '')
```
